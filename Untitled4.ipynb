```python
import scipy.stats as sc
#Подсчет выборочных средних, исправленных дисперсий для двух выборок
x = [1,0,-1,1]
y = [1,1]
xSR = sum(x)/len(x)
ySR = sum(y)/len(y)
xlist = []
ylist = []
for i in x:
    g = (i - xSR)**2
    xlist.append(g)
xDisp = sum(xlist)/(len(xlist)-1)
for j in y:
    k = (j - ySR)**2
    ylist.append(k)
yDisp = sum(ylist)/(len(ylist)-1)
```

```python
#Подсчет для нормального распред - когда по X ищем вероятность 
norm = sc.norm.cdf(0.02)
norm
```

```python
#Подсчет для нормального распред - когда по вероятности ищем X
normX = sc.norm.ppf(0.67)
normX
```

```python
#Подсчет для хи-квадрат: подставляем первым аргументом перцентиль, вторым - степени свободы
chi = sc.chi2.ppf(0.99,1)
chi
```

```python
#Подсчет стьюдента, аналогичная расстановка 
st = sc.t.ppf(0.995,1)
st
```

```python
#Подсчет F: перцентиль - первый аргумент, dfn - второй аргумент (m степеней свободы), dfd - третий арг (n ст.св.)
f = sc.f.ppf(0.99,1,1)
f
```
